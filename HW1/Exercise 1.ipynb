```python
# Create 3 dataframes
dfs = [pd.DataFrame(columns=['country','new cases','new deaths']) for i in range(3)]
```

```python
# Read Guinea data

import os

guinea_dir = "Data/ebola/guinea_data/"
guinea_df = dfs[0]
for file in os.listdir(guinea_dir):
    daily = pd.read_csv(guinea_dir + file, parse_dates=['Date'])
    date = daily.Date[0]
    new_deaths = daily[(daily.Description=='New deaths registered') | (daily.Description=='New deaths registered today')]
    new_deaths.index = new_deaths.Date
    new_deaths_count = int(new_deaths.loc[date,'Totals'])
    new_cases = daily[daily.Description=='New cases of confirmed']
    new_cases.index = new_cases.Date
    new_cases_count = int(new_cases.loc[date,'Totals'])
    guinea_df.loc[date] = ['Guinea',new_cases_count, new_deaths_count]
    
guinea_df.head() 
```

```python
# Read Liberia data

liberia_dir = "Data/ebola/liberia_data/"
liberia_df = dfs[1]
for file in os.listdir(liberia_dir):
    daily = pd.read_csv(liberia_dir + file, parse_dates=['Date']).fillna(0) 
    date = daily.Date[0]
    daily.index = daily.Variable
    new_deaths_count = int(daily.loc['Newly reported deaths','National'])
    new_cases_count = int(daily.loc['New case/s (confirmed)','National'])
    liberia_df.loc[date] = ['Liberia', new_cases_count, new_deaths_count]
```

```python
# Read Sierra Leone data

sl_dir = "Data/ebola/sl_data/"
sl_df = dfs[2]
for file in os.listdir(sl_dir):
    daily = pd.read_csv(sl_dir + file, parse_dates=['date']).fillna(0)
    date = daily.date[0]
    daily.index = daily.variable
    new_cases_count = int(daily.loc['new_confirmed', 'National'])
    new_deaths_count = int(daily.loc['etc_new_deaths', 'National'])
    sl_df.loc[date] = ['Sierra Leone',new_cases_count, new_deaths_count]
```

```python
# Create new dataframes with hierarchical index
new_df = []
for df in dfs:
    df.index.name = 'date'
    new_df.append(df.set_index(['country', df.index]))

# Concatenate the dataframes row-wise
final_df = pd.concat(new_df, axis=0)
final_df
```

At this point we compute the daily average per month of new cases and deaths for each country.
